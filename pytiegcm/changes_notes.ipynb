**Below we show a plot of the composition of the upper atmosphere... the following (Notes from Eric Sutton) should be taken into account:**

**The plot shows some of the issues regarding the diffusive equilibrium extrapolation (for height, we are only using ZG here, which is a simplified version of what we need to have (include ZGMID)):**
1. N2 is inaccurate in the upper 4 pressure levels due to numerical issues
2. The upper most level is not to be trusted for DEN, Z, ZG, ZGMID, (and Eric also thinks) TN due to being too close to the imposed upper boundary conditions. Also there was a bug introduced in v2.0 with the upper most level calculation for ZG and/or ZGMID

**...with these issues in mind, we can see that it's better to start the diffusive equilibrium extrapolation around lev=25, rather than lev=29**
